```python
# Prepare the Data
import torch
from transformers import TapasTokenizer, TapasForQuestionAnswering
import pandas as pd

tsv_path = "your_path_to_the_tsv_file"
table_csv_path = "your_path_to_a_directory_containing_all_csv_files"


class TableDataset(torch.utils.data.Dataset):
    def __init__(self, data, tokenizer):
        self.data = data
        self.tokenizer = tokenizer

    def __getitem__(self, table_key):
        table = pd.read_parquet(f"hf://datasets/cardiffnlp/databench/data/{table_key}/all.parquet").head(250)# be sure to make your table data text only
        encoding = self.tokenizer(
            table=table,
            queries=item.question,
            answer_coordinates=item.answer_coordinates,
            answer_text=item.answer_text,
            truncation=True,
            padding="max_length",
            return_tensors="pt",
        )
        # remove the batch dimension which the tokenizer adds by default
        encoding = {key: val.squeeze(0) for key, val in encoding.items()}
        # add the float_answer which is also required (weak supervision for aggregation case)
        encoding["float_answer"] = torch.tensor(item.float_answer)
        return encoding

    def __len__(self):
        return len(self.data)
```

# EVALUATION

```python
import os
import shutil
import pandas as pd
import torch
from transformers import TapasTokenizer, TapasForQuestionAnswering
from IPython.display import display, HTML

# Create necessary directories
os.makedirs('results/sqa/model', exist_ok=True)

# Initialize model and tokenizer from Hugging Face
model_name = "google/tapas-base-finetuned-wtq"
model = TapasForQuestionAnswering.from_pretrained(model_name)
tokenizer = TapasTokenizer.from_pretrained(model_name)

# Convert table to tokenizer-friendly format
def convert_table_to_inputs(table_data, queries):
    table = pd.DataFrame(table_data.values.tolist(), columns=table_data.columns)
    table = table.map(str)

    # Tokenize the inputs using the Tapas tokenizer
    inputs = tokenizer(
        table=table,
        queries=queries,
        padding="max_length",
        truncation=True,
        return_tensors="pt"
    )
    return inputs, table

def perform_aggregation(aggregation_type, selected_cells):
    # Check if all the cells can be converted to numbers before proceeding.
    if not all(can_convert_to_number(selected_cells)):
        return selected_cells
    if len(selected_cells) == 0:
        return 'None'
    if aggregation_type == 1:
        return sum([float(i) for i in selected_cells])
    elif aggregation_type == 2:
        return len([float(i) for i in selected_cells])
    elif aggregation_type == 3:
        return sum([float(i) for i in selected_cells]) / len(selected_cells)

def can_convert_to_number(lst):
    results = []
    for item in lst:
        try:
            # Try to convert item to a float (this also handles integers)
            float(item)
            results.append(True)
        except ValueError:
            # If conversion fails, mark as False
            results.append(False)
    return results

# Predict the answers
def predict(table_data, queries):
    # Convert table and queries into model inputs
    inputs, table = convert_table_to_inputs(table_data, queries)

    # Run the model for predictions
    outputs = model(**inputs)
    # predicted_answer_coordinates = outputs.logits.argmax(-1)  # Coordinates of answers

    predicted_answer_coordinates, predicted_aggregation_indices = tokenizer.convert_logits_to_predictions(
    inputs, outputs.logits.detach(), outputs.logits_aggregation.detach()
)

    # print(f"inputs: {inputs}")
    # print(f"output logits: {outputs.logits.argmax(-1)}")
    # print(f"outputs: {outputs}")
    # print(f"predicted_answer_coordinates: {predicted_answer_coordinates}")
    # print(f"predicted_aggregation_indices: {predicted_aggregation_indices}")

    # Extract answers
    answers = []
    for i in range(len(predicted_answer_coordinates)):
        if len(predicted_answer_coordinates[i]) == 1:
            # only a single cell:
            answers.append(table.iat[predicted_answer_coordinates[i][0]])
        else:
            # multiple cells
            cell_values = []
            for coordinate in predicted_answer_coordinates[i]:
                cell_values.append(table.iat[coordinate])
            result = perform_aggregation(predicted_aggregation_indices[i], cell_values)
            answers.append(result)

    # Display the table
    # display(HTML(table.to_html(index=False)))

    # Print the query and answers

    for query, answer in zip(queries, answers):
        print(f"> {query}")
        print(f"Answer: {answer}")

    return predicted_answer_coordinates, answer


# Example usage:
table_data = pd.DataFrame({
    "Name": ["Alice", "Bob", "Charlie"],
    "Age": [24, 27, 22],
    "Country": ["USA", "Canada", "UK"]
})

queries = ["What is the age of Alice?", "Who is 27 years old?"]

predicted_answer_coordinates = predict(table_data, queries)
```

```python
def generate_predictions(qa_dataset, max_rows=100):
    predictions = []

    for idx, row in enumerate(qa_dataset):
        print(f"Row Number{idx}")
        question = row['question']
        table_key = row['dataset']  # Get the dataset associated with the question

        df = pd.read_parquet(f"hf://datasets/cardiffnlp/databench/data/{table_key}/all.parquet").head(250)
        queries = [question]

        # print(df)

        predicted_answer_coordinates, answer = predict(df, queries)

        predictions.append(answer)
    return predictions
```
